```python
import pandas as pd
df = pd.read_csv('olympics.csv',index_col=0, skiprows=1)
df.head()
```

```python
df
for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]}, inplace=True)
```

```python
df.head()
```

```python
name_id = df.index.str.split('\s\(')
name_id
```

```python
df.index = name_id.str[0]
df['ID'] = name_id.str[1].str[:3]
```

```python
df = df.drop('Totals')
```

# What is the first country in df?

```python
def answer_zero():
    return df.iloc[0]

answer_zero()
    
```

### Which country has won the most gold medals in summer games?

```python
def answer_one():
    return df['Gold'].idxmax()
answer_one()
```

### Which country had the biggest difference between their summer and winter gold medal counts?

```python
def answer_two():
    return (abs(df['Gold']-df['Gold.1'])).idxmax()
answer_two()
```

### Which country has the biggest difference between their summer gold medal counts and winter gold medal counts relative to their total gold medal count?

```python
def answer_three():
    return ((df['Gold']-df['Gold.1'])/(df['Gold']+df['Gold.1'])).idxmax()
answer_three()
```

##### Write a function that creates a Series called "Points" which is a weighted value where each gold medal (Gold.2) counts for 3 points, silver medals (Silver.2) for 2 points, and bronze medals (Bronze.2) for 1 point. The function should return only the column (a Series object) which you created, with the country names as indices.

```python
def answer_four():
    df['Points'] = df['Gold.2']*3+df['Silver.2']*2+df['Bronze.2']*1
    return df['Points']
answer_four()
```

### Which state has the most counties in it? 

```python
def answer_five():
    return census_df.groupby('STNAME')['COUNTY'].count()
answer_five()
```

### Only looking at the three most populous counties for each state, what are the three most populous states

```python
def answer_six():
    return list(census_df.set_index(['CTYNAME']).sort_values(['CENSUS2010POP'],ascending=False).head(3).index)
answer_six()
```

### Which county has had the largest absolute change in population within the period 2010-2015? (Hint: population values are stored in columns POPESTIMATE2010 through POPESTIMATE2015, you need to consider all six columns.)

```python
def answer_seven():
    return (census_df[['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015','STNAME','CTYNAME']].set_index(['STNAME','CTYNAME']).max(axis=1)-census_df[['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015','STNAME','CTYNAME']].set_index(['STNAME','CTYNAME']).min(axis=1)).idxmax()[0]
answer_seven()
```

### Create a query that finds the counties that belong to regions 1 or 2, whose name starts with 'Washington', and whose POPESTIMATE2015 was greater than their POPESTIMATE 2014.

This function should return a 5x2 DataFrame with the columns = ['STNAME', 'CTYNAME'] and the same index ID as the census_df (sorted ascending by index).

```python
def answer_eight():
    return census_df[((census_df['REGION']==1)|(census_df['REGION']==2))
                    & (census_df['CTYNAME'].str.split().str[0]=='Washington')
                    & (census_df['POPESTIMATE2015']>census_df['POPESTIMATE2014'])][['STNAME','CTYNAME']]
answer_eight()
```
